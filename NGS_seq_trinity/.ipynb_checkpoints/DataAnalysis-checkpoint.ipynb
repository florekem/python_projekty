```python
# slownik, gdzie kazdy dataframe jest wartoscia klucza o tej samej nazwie, przez co moge dodac te nazwe (string)
# do dataframe'a. i w petli zmiana dla wszystkich dataframe'mow

data_frame_dict = {'PBS_vs_MIM_deseq' : PBS_vs_MIM_deseq, 'IVF_vs_MIM_deseq' : IVF_vs_MIM_deseq, 'PBS_vs_IVF_deseq' : PBS_vs_IVF_deseq, 'PBS_vs_MIM_edgeR' : PBS_vs_MIM_edgeR, 'IVF_vs_MIM_edgeR' : IVF_vs_MIM_edgeR, 'PBS_vs_IVF_edgeR' : PBS_vs_IVF_edgeR}

for string,data_frame in data_frame_dict.items():
    data_frame.reset_index(inplace=True)
    data_frame.rename(columns={'index' : 'query_name'}, inplace=True)
    data_frame['data_frame'] = string
```
